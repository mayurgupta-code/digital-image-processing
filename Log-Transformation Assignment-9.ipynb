# Log-Transformation Assignment-9

- imports and setup

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

- image read 
- and convert image to grayscale

```python
img = cv2.imread('tree.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

- Apply log transformation to the grayscale image

```python
log_transformed = np.log(1 + gray)

log_transformed = (255 * (log_transformed / np.max(log_transformed))).astype(np.uint8)
```

- plot of original image vs histogram
- plot of log transformed image vs its histogram

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
plt.subplot(2,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original Image")

plt.subplot(2,2,2)
plt.hist(gray.ravel(), bins=256, range=(0, 256))
plt.title('Histogram of Original Image')
plt.xlim([0, 256])

plt.subplot(2,2,3)
plt.imshow(log_transformed, cmap='gray')
plt.title("Log Transformed Image")

plt.subplot(2,2,4)
plt.hist(log_transformed.ravel(), bins=256, range=(0, 256))
plt.title('Histogram of Log Transformed Image')
plt.xlim([0, 256])

plt.show()
```
