# Bit-Plane Slicing Assignment-5

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def bit_slice(data):
    images = []
    div = 128
    for i in range(0,8,1):
        temp = data&(128>>i)
        images.append(temp)
    return images
```

```python
import matplotlib.image as mpimg

img = mpimg.imread('man-looking-his-watch.jpg')
sliced = bit_slice(img)
```

```python
def plots(data):
    plt.subplots(figsize=(20,10))
    for i in range(0,6,1):
        plt.subplot(2,3,i+1)
        plt.imshow(data[i])
```

```python
plots(sliced)
```

```python
img2 = mpimg.imread('mountains.jpg')
plots(bit_slice(img2))
```
