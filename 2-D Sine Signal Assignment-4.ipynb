- generating sine signal

```python
def sineGenerator(x, y):
    return np.sin(2*math.pi*x + 2*math.pi*y)
    
```

```python
X, Y = np.meshgrid(x, y)
z = sineGenerator(X,Y)
```

```python
z
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, z, 50, cmap='binary')
```
