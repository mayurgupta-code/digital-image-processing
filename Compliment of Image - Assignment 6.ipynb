# Compliment of Image - Assignment 6

- imports and setup

```python
import cv2
import copy
import matplotlib.pyplot as plt
import numpy as np
```

- defining imageCompliment function - to Take the Compliment of the image
- we can take compliment of the image by subtracting each pixel of the image by 255

- np.abs() is a function that returns the absolute value (magnitude) of the input array. The absolute value of a number is its distance from zero, regardless of its sign.

```python
img = cv2.imread('flowers.jpg', 0)


# Compute the complement of the image
img_complement = np.abs(img - 255)

```

- plot the original and resultant image using matplotlib

```python
plt.figure(figsize=(15,15))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original")

plt.subplot(1, 2, 2)
plt.imshow(img_complement)
plt.title("Compliment")
plt.show()
```
