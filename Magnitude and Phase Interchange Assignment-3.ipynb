```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy.fft as fft
import cmath
```

```python
img = mpimg.imread('buildings.jpg')
plt.imshow(img)
```

```python
transform = fft.fft2(img)
```

```python
plt.imshow(abs(transform))
```

```python
plt.imshow(np.angle(transform))
```

```python
img2 = mpimg.imread('flowers.jpg')
t2= fft.fft2(img2)
plt.subplots(figsize=(20,10))
plt.subplot(2,3,1)
plt.imshow(img2)
plt.subplot(2,3,2)
plt.imshow(abs(t2).astype(np.uint8))
plt.subplot(2,3,3)
plt.imshow(np.angle(t2).astype(np.uint8))
```

```python
img3 = mpimg.imread('bridge.jpg')
t3 = fft.fft2(img3)
plt.subplots(figsize=(20,10))
plt.subplot(2,3,1)
plt.imshow(img3)
plt.subplot(2,3,2)
plt.imshow(abs(t3).astype(np.uint8))
plt.subplot(2,3,3)
plt.imshow(np.angle(t3).astype(np.uint8))
```

```python
sum(np.angle(t3))
```

```python
tempMag = np.abs(t2)
tempMag = tempMag.astype(np.uint8)
temp_phase = np.angle(t3)
temp_phase = temp_phase.astype(np.uint8)
t4 = tempMag*temp_phase
img4 = fft.ifft2(t4)
img4 = img4.astype(np.uint8)
plt.imshow(img4)
```
