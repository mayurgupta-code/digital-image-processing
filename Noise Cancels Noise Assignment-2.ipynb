# Noise Cancels Noise Assignment-2

- imports and image read

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

img = mpimg.imread('tree.jpg')
```

- show image

```python
plt.imshow(img)
```

```python
img = img.astype(np.float32)
```

- taking mean and std of image

```python
mean = np.mean(img)
std = np.std(img)
```

- adding noise to the image

```python
noise = np.random.normal(0,0.5*std,img.shape)
noisy_image = img+noise
```

```python
noisy_image = noisy_image.astype(np.uint8)
```

```python
plt.imshow(noisy_image)
```

```python
def get_noisy_image(img,alpha):
    mean = np.mean(img)
    std = np.std(img)
    noise = np.random.normal(0,alpha*std,img.shape)
    noisy_image = img+noise
    return noisy_image
```

```python
alpha = 0.6
noisy_set = []
for i in range(0,100,1):
    temp_img = get_noisy_image(img,alpha)
    noisy_set.append(temp_img)
```

```python
sum_image = np.zeros((img.shape))
for i in range(0,100,1):
    sum_image +=noisy_set[i]
resultant = sum_image//100
```

```python
resultant= resultant.astype(np.uint8)
plt.imshow(resultant)
```
