# Arithmetic Mean Filtering Assignment-10

- imports and setup
- done with the help of python-opencv

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('sunset.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
b, g, r = cv2.split(img)
```

- define 3x3 and 5x5 kernel matrices for arithmetic mean filter

```python
k1 = np.ones((3, 3), np.float32) / 9
k2 = np.ones((5, 5), np.float32) / 25

kernel3 = k1
kernel5 = k2

# apply arithmetic mean filter to the grayscale image using the 3x3 kernel
gray_3 = cv2.filter2D(gray_img, -1, k1)
gray_5 = cv2.filter2D(gray_img, -1, k2)

# apply arithmetic mean filter to the red channel using the 3x3 kernel
red_3 = cv2.filter2D(r, -1, kernel3)
red_5 = cv2.filter2D(r, -1, kernel5)

# apply arithmetic mean filter to the green channel using the 3x3 kernel
green_3 = cv2.filter2D(g, -1, kernel3)
green_5 = cv2.filter2D(g, -1, kernel5)

# apply arithmetic mean filter to the blue channel using the 3x3 kernel
blue_3 = cv2.filter2D(b, -1, kernel3)
blue_5 = cv2.filter2D(b, -1, kernel5)
```

- plot the images for grayscale

```python
plt.figure(figsize=(15, 15))

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(2, 2, 3)
plt.imshow(gray_3, cmap='gray')
plt.title('Grayscale Image 3x3')

plt.subplot(2, 2, 4)
plt.imshow(gray_5, cmap='gray')
plt.title('Grayscale Image 5x5')
```

- plot the images for red

```python
plt.figure(figsize=(15, 15))

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(r, cmap='Reds')
plt.title('Red Channel')

plt.subplot(2, 2, 3)
plt.imshow(red_3, cmap='Reds')
plt.title('Red Channel 3x3')

plt.subplot(2, 2, 4)
plt.imshow(red_5, cmap='Reds')
plt.title('Red Channel 5x5')
```

- plot the images for green

```python
plt.figure(figsize=(15, 15))

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(g, cmap='Greens')
plt.title('Green Channel')

plt.subplot(2, 2, 3)
plt.imshow(green_3, cmap='Greens')
plt.title('Green Channel 3x3')

plt.subplot(2, 2, 4)
plt.imshow(green_5, cmap='Greens')
plt.title('Green Channel 5x5')
```

- plot the images for blue

```python
plt.figure(figsize=(15, 15))

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(b, cmap='Blues')
plt.title('Blue Channel')

plt.subplot(2, 2, 3)
plt.imshow(blue_3, cmap='Blues')
plt.title('Blue Channel 3x3')

plt.subplot(2, 2, 4)
plt.imshow(blue_5, cmap='Blues')
plt.title('Blue Channel 5x5')

plt.show()
```
